Viết hàm tìm phần tử có số
lần xuất hiện nhiều nhất
trong danh sách. Nếu có
nhiều phần tử cùng số lần,
trả về phần tử xuất hiện
trước

```python
t=list(map(int,input().split()))
def dem(n,t):
    d=0
    for i in t:
        if i==n:
            d=d+1
    return d
max=dem(t[0],t)
j=t[0]
for i in t:
    if dem(i,t)>max:
        max=dem(i,t)
        j=i
print(j,'xuat hien',max)



```

Cho một danh sách các số
nguyên nums, hãy tìm độ
dài của dãy con tăng
nghiêm ngặt dài nhất
(không nhất thiết phải liền kề)

```python
t=list (map(int,input().split(',')))
max=0
def timmax(h):
    max=h[0]
    for i in h :
        if i >max :
            max=i
    return max

for i in range(len(t)) :
    d=0
    h=[]
    h.append(t[i])
    for j in range(i,len(t)):
        if t[j]>=timmax(h):
            d=d+1
            h.append(t[j])
    if d>max :
        max=d
print(max)
```

Cho một danh sách các số
nums, với mỗi phần tử
nums[i], hãy đếm xem có
bao nhiêu phần tử bên phải
nhỏ hơn nums[i].
Trả về danh sách các số
đếm.
Efficient solution: using
heap or binary search tree

```python
t=list (map(int,input().split(',')))
k=[]
for i in range(len(t)):
    d=0
    for j in range(i+1,len(t)):
        if t[j]<t[i]:
            d=d+1
    k.append(d)
print(k)
```

Cho một danh sách các số
nguyên nums và một số
nguyên m, hãy tìm tổng
mảng con lớn nhất theo
modulo m.
Tức là, với bất kỳ mảng
con nums[i:j] nào, hãy tính
sum(nums[i:j]) % m và trả
về kết quả lớn nhất có thể.
Gợi ý:
Duy trì tổng tiền tố (prefix
sum) mod m và sử dụng
tìm kiếm nhị phân để tìm
tổng tiền tố nhỏ nhất chỉ
lớn hơn tổng hiện tại một
chút (để tối đa hóa hiệu số
bao quanh).


```python
t=list (map(int,input().split(',')))
m=int(input())
import bisect

def max_sub_mod(a, m):
    ans = 0
    cur = 0
    p = []
    
    for x in a:
        cur = (cur + x) % m
        ans = max(ans, cur)
        
        i = bisect.bisect_right(p, cur)
        if i < len(p):
            ans = max(ans, (cur - p[i] + m) % m)
            
        bisect.insort(p, cur)
        
    return ans
print(max_sub_mod(t, m))
```
