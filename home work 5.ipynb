Viết hàm kiểm tra một số có phải số Amstronhg hay không? Hàm trả về kiểu bool, có tham số là một số
nguyên x(x=tổng các chữ số mũ lũy thừa số phần tử của số vd:153=1^3+5^3+3^3)

```python
n=int(input())
def ams(n):
    l=len(str(n))
    k=n
    t=0
    while n>0 :
        m=n%10
        t=t+m**l
        n=n//10
    if t==k :
            return True
    else :
            return False
print(ams(n))
```

VIết hàm kiểm tra một số có phải số kỳ diệu hay không, xử lý trên dữ liệu kiểu số nguyên(Trong toán học và lập trình, khái niệm "số kỳ diệu" (Magic Number) thường được hiểu là số mà khi ta tính tổng các chữ số của nó, rồi lại tiếp tục tính tổng các chữ số của kết quả đó cho đến khi chỉ còn một chữ số, nếu kết quả cuối cùng bằng 1 thì đó là số kỳ diệu.)

```python
n=int(input())
def skd(n):    
    while n>9:    
        t=0
        while n>0:
            m=n%10
            t=t+m
            n=n//10
        n=t
    if n==1:
        return True
    else :
        return  False
print(skd(n))
```

Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).


```python
x1,y1,x2,y2,x3,y3=map(float,input().split())
def goc(x1,y1,x2,y2,x3,y3):
    import math
    u1, u2 = x1 - x2, y1 - y2
    v1, v2 = x3 - x2, y3 - y2
    t= u1 * v1 + u2 * v2
    ba = (u1**2 + u2**2)**0.5
    bc = (v1**2 + v2**2)**0.5
    if ba==0 or bc==0 :
        return 'k ton tai goc'
    else :
        cs=t/(ba*bc)
        if -1<=cs<=1 :
            rad=math.acos(cs)
            do=math.degrees(rad)
            return do
        else :
            return 'kton tai goc'
print(goc(x1,y1,x2,y2,x3,y3))
    
```

Viết hàm trả về giá trị lớn nhất và nhỏ nhất của ba số a,b,c dưới dạng list [max, min].


```python
a=list(map(float,input().split()))
def tim(a) :
    max=a[0]
    min=a[0]
    for i in range(len(a)):
        if a[i]>max:
            max=a[i]
        if a[i]<min :
            min=a[i]
    return max,min
print(tim(a))
```

Viết hàm tính hệ số nhị thức C(n,k). tính n!/(k!*(n-k)!)

```python
n,k=map(int,input().split())
def nhithuc(n,k):  
    if n>=k>0 :
        t=1
        t1=1
        t2=1
        for i in range(1,n+1):
            t=t*i
        for i in range(1,k+1):
            t1=t1*i
        for i in range(1,n-k+1):
            t2=t2*i
        a=t/(t1*t2)
        return a
    else :
        return 'nhap lai'
print(nhithuc(n,k))
```

Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau
Nếu n ≤ 0 trả về 0 (tổng các số trong đoạn có các chữ số đôi một khác nhau)

```python
n=int(input())
def sum(n):
    if n>0:    
        t=0
        for i in range(1,n+1):
            x=str(i)
            if len(x)==len(set(x)):
                t=t+i
        return t
    else :
        return 0
print(sum(n))
```
