Xây dựng lớp Cylinder (hình trụ) gồm: bán kính và chiều cao
(kiểu số thực) với hai phương thức:tính diện tích bề mặt (diện
tích toàn phần) và thể tích của hình trụ
Đầu vào
Đâu vào từ bàn phím gồm một dòng duy nhất chứa bán kính và
chiều cao của hình trụ, cách nhau bởi một dấu cách.
Đầu ra
In ra màn hình diện tích bề mặt và thể tích của hình trụ (sử
dụng lại những hàm vừa viết, π=3.14).
(Kết quả làm tròn đến 2 chữ số sau dấu phẩy). 


```python
class Cylinder:
    def __init__(self,d,r):
        self.d=d
        self.r=r
    def dt(self):
        return (self.r**2)*3.14*2+2*3.14*self.d*self.r 
    def tt(self):
        return (self.r**2)*3.14*self.d
d,r=map(float,input().split())
t=Cylinder(d,r)
print(t.dt())
```

area() trả về diện tích.
perimeter() trả về chu vi.
scale(k) phóng to hình chữ nhật theo hệ số k (nhân cả w và h),
không trả về gì.
Input
w h k (số thực dương)
Output
area perimeter sau khi phóng to (làm tròn 2 chữ số thập phân). 

```python
class Rec:
    def __init__(self,x,y):
        self.x=x
        self.y=y
    def area(self):
        return self.x*self.y
    def per(self):
        return (self.x+self.y)*2
    def scale(self,k) :
        self.x=x*k
        self.y=y*k
x,y,k=map(float,input().strip().split())
t=Rec(x,y)
print(t.area())
t.scale(k)
print(t.area())
```
