### Bài 2: Tìm số nhỏ nhất của một danh sách các số nguyên. KHONG DUNG HAM SAN CO CUA LIST

```python
# lst = list(map(int, input().split()))

def min_of_list (lst):
    if len(lst)>0:
        min=lst[0]
        for i in lst:
            if i<min:
                min=i
        return min
    else :
        return None

assert min_of_list([]) == None
assert min_of_list([23, 67, 900, -10]) == -10
assert min_of_list([575, 939, 697, 138, 183, 386, 147, 695, 622, 289]) == 138

print("Pass all Tests!")
```

### Bài 2b: Tìm số nhỏ nhất của 1 list và vị trí của nó, nếu có nhiều số nhỏ nhất bằng nhau thì lấy chỉ số đầu tiên (nhỏ nhất)
10 0 0 12 34 5 6 7  8 89 0 1
0 1

```python


def min_and_pos(lst):
    if len(lst)>0:
        min=lst[0]
        j=0
        d=0
        for i in lst:
            j=j+1
            if i<min:
                min=i
                d=j

        return (min, d)
    else :
        return None

assert min_and_pos([10, 0, 0, 12, 34, 5, 6, 7, 8, 89, 0, 1]) == (0, 2)
assert min_and_pos([]) == None

print("Pass all Tests!!!")
```

### Bài 3: 
Tìm giá trị trung bình của các số nguyên dương trong một list.
Nếu không có số dương nào thì trả về "invalid"

```python
#Tuan
def Average_lst(lst):
    if not lst:
        return None
    else :
        t=0
        d=0
        for i in lst:
            if i>0:
                t=t+i
                d=d+1
        if d!=0:
            return round(t/d,2)
        else:
            return None
    

assert Average_lst([1, 2, 6, 1, -10, -10]) == 2.50
assert Average_lst([-2, -1, -20, -343]) == None
assert Average_lst([]) == None

print("Accepted!")
```

### Bài 4: Kiểm tra một list số nguyên có phải là đối xứng hay không
[1, 2, 3, 5, 3, 2, 1]

```python
def kt_doi_xung(lst):
    if not lst:
        return None
    else:
        t=lst[::-1]
        return t==lst



assert kt_doi_xung([1, 2, 3, 5, 3, 2, 1]) == True
assert kt_doi_xung([1, 2, 3, 5, 3, 4, 1]) != True
assert kt_doi_xung([]) == None


print("Accepted!!")
```

### Bài 5: Đưa tất cả các số bằng 0 trong list về đầu của list. Các số khác không thay đổi thứ tự trước sau
3 4 6 7 1 9 0 0 23 0 0 34 67 89
0 0 0 0 3 4 6 7 1 9 23 34 67 89

```python
def move_zero_to_first(lst):
    k=[]
    h=[]
    for i in lst:
        if i==0:
            k.append(i)
        else:
            h.append(i)       
    return k+h
    
    

assert move_zero_to_first([3, 4, 0, 0, 23, 0, 0, 34, 67, 89]) == [0, 0, 0, 0, 3, 4, 23, 34, 67, 89]
assert move_zero_to_first([0, 0, 0, 0]) == [0, 0, 0, 0]
assert move_zero_to_first([3, 4, 0, 0, 1, 2, 3]) == [0, 0, 3, 4, 1, 2, 3]

print("accepted!!!")
```

### Bài 6: Xóa tất cả các số giống nhau trong list, giữ lại đúng 1 lần, Các số khác giữ nguyên vị trí trước 
[1, 2, 3, 4, 1, 2, 3, 4, 1, 1, 4, 3] 


```python
def unique_array(lst):
    seen=[]
    if not lst :
        return None
    else:
        for i in lst :
            if i not in seen :
                seen.append(i)
        return seen
       




assert unique_array([1, 2, 3, 4, 1, 2, 3, 4, 1, 1, 4, 3]) == [1,2,3,4]
assert unique_array([1, 1, 1, 1]) == [1]
assert unique_array([]) == None

print("Accepted!!!")
```

### Bài 7: Kiểm tra một list số nguyên có phải là dãy tăng, dãy giảm hay vô danh
1 3 5 6 7 8 9: Tang

9 8 7 6 5 4 0: Giam

1 2 3 7 1 0 2: Vo danh

1 2 2 4 5 7: Vo danh

```python
def check_order_of_list(lst):
    if len(set(lst))==len(lst)    :
        h=sorted(lst)
        k=h[::-1]
        if h==lst :
            return 'Tang'
        elif lst==k:
            return 'Giam'
        else :
            return "Vo danh"
    else:
        return "Vo danh"
assert check_order_of_list([1, 3, 5, 6, 7, 8, 9]) == "Tang"
assert check_order_of_list([9, 8, 7, 6, 5, 4, 0]) == "Giam"
assert check_order_of_list([1, 2, 3, 7, 1]) == "Vo danh"
assert check_order_of_list([7, 6, 5, 4, 3, 3]) == "Vo danh"
print("Pass all tests!!!")
```

### Bài 8: Kiểm tra một danh sách số nguyên có phải là ziczac khong
1 3 2 5 2 8 7 9 2 100: ziczac
34 2 56 7 90 20 30 7 120: ziczac
2 2 34 4 56 7 89 : không phải ziczac


```python
def is_ziczac(lst):
    for i in range(1,len(lst)-1):
        if i%2==1:
            if (lst[i-1]<lst[i] and lst[i+1]<lst[i]) or (lst[i-1]>lst[i] and lst[i+1]>lst[i]):
                continue
            else :
                return False
        else :
            if (lst[i-1]>lst[i] and lst[i+1]>lst[i]) or (lst[i-1]<lst[i] and lst[i+1]<lst[i]):
                continue
            else :
                return False
    return True

assert is_ziczac([1, 3, 2, 5, 0, 7, 6, 8,0]) == True
assert is_ziczac([34, 2, 56, 7, 80, 12, 24, 14]) != False
assert is_ziczac([2, 2, 34, 4, 56, 7, 89]) == False

print("All accepted!")
```

### Bài 9: Tìm và in ra tất cả các vị trí của số nguyên k trong một danh sách số nguyên cho trước. Nếu không có k trong ds thì in ra "-1".
VD: k = 3, [1,3,4,4,3,3,4,3,6,1] => k xuất hiện tại các vị trí (trả về 1 list) [1, 4, 5, 7]
k = 10, [1,3,4,4,3,3,4,3,6,1] => k không xuất hiện trong ds => trả về None

```python
def find_pos_of_k(lst, k):
    h=[]
    for i in range(len(lst)):
        if lst[i]==k:
            h.append(i)
    if not h :
        return None
    else:
        return h


assert find_pos_of_k([1,3,4,4,3,3,4,3,6,1], 3) == [1, 4, 5, 7]
assert find_pos_of_k([1,3,4,4,3,3,4,3,6,1], 10) == None
assert find_pos_of_k([1, 1, 2, 3], 3) == [3]

print("Tests accepted!")
```

VD: "Nguyen Ai Quoc" => ["N", "c"]

```python
def theFirst_Last_character(s):
    if not s :
        return None
    else :
        h=[]
        h.append(s[0])
        h.append(s[-1])
        return h

assert theFirst_Last_character("Nguyen Ai Quoc") == ["N", "c"]
assert theFirst_Last_character("Ton Duc Thang") == ["T", "g"]
assert theFirst_Last_character("") == None

print("passed all tests!")
```

### Bài 11: 
Cho một danh sách các số nguyên có giá trị trong đoạn [0,9]. Liệt kê số lần xuất hiện của các số có mặt trong danh sách. Số nào xuất hiện trước thì liệt kê trước
VD 1 2 1 2 1 1 1 1

output: 
1: 6
2: 2


```python
def list_number_in_lst(lst):
    if not lst :
        return None
    else :    
        k={}
        for i in lst:
            if i not in k :
                d=0
                k[i]=[] 
                for j in lst:
                    if j ==i :
                        d=d+1
                k[i]=d
        return k 

assert list_number_in_lst([1, 2, 1, 1, 2, 1, 1]) == {1: 5, 2: 2}
assert list_number_in_lst([1, 2, 1, 1, 2, 1, 1, 3, 3, 4, 2, 1, 2, 3]) == {1: 6, 2: 4, 3: 3, 4: 1}
assert list_number_in_lst([]) == None
assert list_number_in_lst([4, 3, 1, 2, 1, 1, 2, 1, 1])== {4: 1, 3: 1, 1: 5, 2: 2}

print("Pass all tests!")
```

### Bài 12: 
Cho một chuỗi kí tự chỉ gồm các kí tự chữ cái thường 'a'....'z'. Hãy liệt kê số lần xuất hiện của các kí tự có mặt trong chuỗi theo thứ tự tăng dần.
VD: "abcabcccc"
output:
a : 2
b : 2
c : 5

```python
def count_character_of_string(lst):
    if not lst :
        return None
    else :    
        k={}
        for i in lst:
            if i not in k :
                d=0
                k[i]=[] 
                for j in lst:
                    if j ==i :
                        d=d+1
                k[i]=d
        return k 

assert count_character_of_string("abcabccc") == {'a': 2, 'b': 2, 'c': 4}
assert count_character_of_string("fbdfgfhsgfjsghghg") == {'f': 4, 'b': 1, 'd': 1, 'g': 5, 'h': 3, 's': 2, 'j': 1} 

print("Passed!")
```
