Cho một mảng số nguyên không âm. Gọi mảng đó là "mảng đẹp" nếu
mỗi phần tử x trong mảng xuất hiện đúng x lần.
Kiểm tra xem mảng có phải là mảng đẹp hay không. Nếu không hãy xóa
đi y số (y là nhỏ nhất có thể) để mảng còn lại thành mảng đẹp

```python
t=list(map(int,input().split()))
t1=t
t2=set(t)
for i in range(len(t2)):
    d=0
    for j in range(len(t)):
        if t[j]==t[i]:
            d=d+1
    if d>t[i]:
        h=d-t[i]
        for ii in range(h):
            t1.remove(t[i])
    elif d<t[i]:
        for ii in range(d):
            t1.remove(t[i])
print(t1)
```

Gặp 5: Vì đã hết lượt xóa (rem = 0), ta không so sánh nữa, cứ thế đẩy vào Stack.

Gặp 3: Tương tự, đẩy vào Stack.

Stack hiện tại: [9, 8, 9, 5, 3]


```python
def tim_so_lon_nhat(s, k):
    n = len(s)
    rem = n - k
    stack = []
    
    for c in s:
        while rem > 0 and stack and stack[-1] < c:
            stack.pop()
            rem -= 1
        stack.append(c)
        
    return "".join(stack[:k])

s =input()
k =int(input())
print(tim_so_lon_nhat(s, k))
```

Bà̀i toán Tìm dãy các hợp số dài nhất liên tiếp không chứa số nguyên
nào <= N (N là số nguyên dương cho trước). Nếu có nhiều dãy cùng độ
dài thì lấy dãy nhỏ nhất (gặp đầu tiên từ trái qua)
hợp số xuất hiện giữ hai số nguyên tố liền kề
Cách giải tìm hai số nguyên tố lớn hơn N gần nhất

VD: giữ 7 và 11 có dãy hợp số 8 9 10 

```python
def la_so_nguyen_to(num):
    if num < 2:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True

def tim_day_hop_so_dai_nhat(n):
    p1 = n + 1
    while not la_so_nguyen_to(p1):
        p1 += 1
        
    p2 = p1 + 1
    while not la_so_nguyen_to(p2):
        p2 += 1
        
    return list(range(p1 + 1, p2))

N = int(input())
print(tim_day_hop_so_dai_nhat(N))
```

Tìm Dãy con tăng dần liên tục dài nhất của một danh sách nhập vào từ
bàn phím. Lấy dãy đầu tiên nếu có nhiều dãy có số phần tử bằng nhau

```python
def tim_day_con_tang_lien_tuc(arr):
    if not arr:
        return []

    max_len = 1
    best_start = 0
    
    curr_len = 1
    curr_start = 0

    for i in range(1, len(arr)):
        if arr[i] > arr[i-1]:
            curr_len += 1
        else:
            if curr_len > max_len:
                max_len = curr_len
                best_start = curr_start
            curr_start = i
            curr_len = 1

    if curr_len > max_len:
        max_len = curr_len
        best_start = curr_start

    return arr[best_start : best_start + max_len]

nhap_chuoi = input()
arr = [int(x) for x in nhap_chuoi.split()]

ket_qua = tim_day_con_tang_lien_tuc(arr)
print( ket_qua)
```
