W5A3 Viết hàm kiểm tra một số có phải là số nguyên tố hay không? 

```python
n=int(input())
def nto(n):
    d=0
    for i in range(1,n+1) :
        if n%i==0 :
            d=d+1
    if d==2 :
        return True
    else :
        return False
print(nto(n))
```

```python
#cach 2
n=int(input())
def nto(n):
    if n<2 :
        return False
    else :
        for i in range(2,int(n**0.5)+1) :
            if n%i==0 :
                return False
        else :
            return True        
print(nto(n))
```

W5A4 VIết hàm kiểm tra một số có phải số hoàn hảo hay không, xử lý trên dữ liệu kiểu số nguyên
(tổng các uosc trừ nó bằng chính nó )

```python
a=int(input())
def hoanhoa(n) :    
    t=0
    for i in range(1,n,1) :
        if a%i==0 :
            t=t+i
    if t==a :
        return True
    else :
        return False
print(hoanhoa(a))
```

Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và trả về vị trí của phần tử đầu tiên có giá trị k trong list số nguyên, nếu không có thì trả về -1 


```python
a=list(map(int,input().split()))
k=int(input())
def timk(a,k):
    for i in range(len(a)):
        if a[i]==k :
           return i+1
    return -1
print(timk(a,k))

```

W5A6 Viết hàm tính giai thừa. Nhập vào số nguyên n, trả về giai thừa của n (n!) 


```python
n=int(input())
def giathua(n):
    t=1
    for i in range(1,n+1):
        t=t*i
    return t
print(giathua(n))
```

Hàm thực hiện tính toán kết quả của phép toán giữa số \( num1 \) và \( num2 \) với toán tử \( operat
\)
Biết các toán tử đơn giản thực hiện phép toán lần lượt là:
Phép trừ \( - \)
Phép cộng \( + \)
Phép chia \( / \)
Phép nhân \( * \)
Kết quả làm tròn đến chữ số thập phân thứ hai.
Đầu vào: Hai số thực và phé

```python
a=list(map(str,input().split()))
def mtih(a) :
    b=float(a[0])
    c=float(a[1])
    d=str(a[2])
    if d=='+':
        return f"{(b+c):.2f}"
    elif d=='-':
        return f"{(b-c):.2f}"
    elif d=='*':
        return f"{(b*c):.2f}"
    elif d=='/':
        return f"{(b/c):.2f}"
print(mtih(a))
```

0 <= |r| < |divisor|.

W5A15 Viết hàm is_right_triangle(a, b, c) trả về True nếu ba số có thể là ba cạnh của tam giác vuông

```python
a,b,c=map(int,input().split())
def ktr(a,b,c):
    if a**2+b**2==c**2 or a**2+c**2==b**2 or b**2+c**2==a**2 :
        return True
    else:
        return False
print(ktr(a,b,c))
```
