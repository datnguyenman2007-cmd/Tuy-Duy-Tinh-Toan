B.16 Nhâp vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma ̣
trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

```python
m,n=map(int,input().split())
t=[]
for i in range(1,n+1):
    #chạy n hàng
    h=[]
    for j in range(1,m+1):
        #chạy j tạo hàng có m phần tử(tạo cột)
        k=int(input())
        if len(str(k))<4:
            d=4-len(str(k))
            k=d*' '+str(k)
        else:
            k=str(k)
        
        h.append(k)
        #tạo hàng
    t.append(h)
    #thêm hàng vào ma trận
#tạo ma trận
for h in t:
    # Nối các chuỗi trong hàng lại với nhau và in
    print(''.join(h))
    #join chỉ nhập kiểu str


```

B.17 Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và
đường chéo phụ của mảng


```python
n=int(input())
t=[]
for i in range(1,n+1):
    k=list(map(int,input().split()))
    t.append(k)
for i in range(n):
    print(t[i][i],end=' ')
print('\0')
for i in range(n):
    print(t[n-i-1][i],end=' ')
```
