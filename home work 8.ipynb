Yêu cầu
Xây dựng lớp Shape với phương thức area().
Xây dựng lớp Rectangle(w, h) và Circle(r) kế thừa từ Shape, ghi đè area().
Lấy π = 3.14.
Input
Dòng 1: loại hình (RECTANGLE hoặc CIRCLE)
Dòng 2: tham số tương ứng.
Output
In ra diện tích với 2 chữ số thập phân.

```python
class shape :
    def area():
        pass
class HCN(shape):
    def __init__(self,d,r):
        self.d=d
        self.r=r
    def area(self):
        return self.d*self.r
       
class TRON(shape):
    def __init__(self,r):
        self.r=r
    def area(self):
        return 3.14*(self.r**2)
n=input().strip()
if n =='HCN':
    d,r=map(int,input().split())
    dt=HCN(d,r)
    print(dt.area())
if n =='TRON':
    r=int(input())
    dt=TRON(r)
    print(dt.area())
```

Yêu cầu
Xây dựng lớp Polygon với phương thức perimeter().
Xây dựng lớp Triangle(a, b, c) kế thừa từ Polygon và ghi đè perimeter() trả về a+b+c.
Input
Một dòng chứa 3 số nguyên a b c.
Output
In ra chu vi tam giác. 

```python
class Polygon:
    def perimeter():
        pass
class Triangle(Polygon):
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
    def perimeter(self):
        return self.a+self.b+self.c
a,b,c=map(int,input().split())
k=Triangle(a,b,c)
print(k.perimeter())

       
```

balance rate
Output
Số dư sau khi cộng lãi, in 2 chữ số thập phân

Yêu cầu
Xây dựng lớp Matrix2x2(a, b, c, d) biểu diễn ma trận:
[a b]
[c d]
Viết phương thức trace() trả về a + d và determinant() trả về ad - bc.
Input
Một dòng chứa 4 số nguyên a b c d.
Output
In ra trace và determinant trên cùng một dòng. 


```python
class Matr :
    def __init__(self,a,b,c,d):
        self.a=a
        self.b=b
        self.c=c
        self.d=d
    def tr(self):
        return self.a+self.d
    def dete(self):
        return self.d*self.a-self.b*self.c
a,b,c,d=map(float,input().strip().split())
t=Matr(a,b,c,d)
print(t.tr())
print(t.dete())
        
```

Yêu cầu
Xây dựng lớp Product(name, price).
Xây dựng lớp DiscountProduct(name, price, percent) kế thừa từ Product, ghi đè phương thức
final_price() để trả về giá sau giảm.
Công thức: price * (100 - percent) / 100
Input
name price percent
Output
In ra giá cuối cùng với 2 chữ số thập phân

```python
class Pro :
    def __init__(self,price,name):
        self.a=price
        self.b=name
class Dis(Pro):
    def __init__(self,name, price, per,):
        super().__init__(price, per)
        self.c=per
    def final(self):
        return self.a*(100-self.c)/100
a,b,c=map(int,input().split())
t=Dis(a,b,c)
print(t.final())
```

Yêu cầu
Xây dựng lớp Time(h, m, s) với phương thức to_seconds() đổi thời gian sang tổng số giây.
Input
h m s
Output
Tổng số giây

```python
class time:
    def __int__(self,h,m,s):
        self.h=s
        self.m=m
        self.s=s
    def second(self):
        return self.h*3600+self.m*60+self.s
h,m,s=map(int,input().split())
t=time(h,m,s)
print(t.second())
```
